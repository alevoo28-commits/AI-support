#### 5. **VectorStoreRetrieverMemory (FAISS)**
- **Prop√≥sito**: Memoria sem√°ntica a largo plazo
- **Implementaci√≥n**: `self.vector_memory` con FAISS
- **Uso**: B√∫squeda por similitud de contenido

### üîç **RAG Principal con FAISS**

#### **Implementaci√≥n Completa:**
```python
# Cada agente tiene su propio vectorstore FAISS
self.vectorstore_rag = FAISS.from_documents(chunks, self.embeddings)

# B√∫squeda sem√°ntica implementada
def buscar_contexto_faiss(self, consulta: str) -> str:
    docs = self.vectorstore_rag.similarity_search(consulta, k=3)
    contexto = "\n\n".join([doc.page_content for doc in docs])
    return contexto
```

#### **Caracter√≠sticas FAISS:**
- ‚úÖ **Material vectorizado**: `soporte_informatica.txt` con embeddings
- ‚úÖ **B√∫squeda sem√°ntica**: `similarity_search()` por consulta
- ‚úÖ **Contexto relevante**: Top 3 chunks m√°s similares
- ‚úÖ **Especializaci√≥n**: Material espec√≠fico por agente
- ‚úÖ **Integraci√≥n**: FAISS en prompts de agentes

#### **Flujo FAISS:**
```
Consulta ‚Üí similarity_search() ‚Üí Contexto relevante ‚Üí Prompt ‚Üí Respuesta
```

## üìè M√©trica de Consistencia

### ‚úÖ **Definici√≥n de Consistencia**

La **consistencia** mide la capacidad del sistema multi-agente para proporcionar respuestas similares y coherentes ante consultas iguales o sem√°nticamente equivalentes en diferentes momentos o sesiones.

### üéØ **Importancia de la Consistencia**

- **Confiabilidad**: Los usuarios esperan respuestas predecibles ante preguntas similares
- **Estabilidad**: El sistema debe mantener su comportamiento a lo largo del tiempo
- **Credibilidad**: Respuestas inconsistentes erosionan la confianza del usuario
- **Calidad**: La consistencia es indicador de un sistema bien configurado

---

### üìä **Metodolog√≠a de Evaluaci√≥n de Consistencia**

#### 1. **Pruebas de Regresi√≥n**

Se ejecutan consultas id√©nticas en diferentes sesiones y se comparan las respuestas:

```python
# Ejemplo de test de consistencia
consulta_prueba = "¬øC√≥mo puedo mejorar el rendimiento de mi computadora?"

# Ejecutar 10 veces en diferentes sesiones
resultados = []
for i in range(10):
    respuesta = sistema.procesar_consulta(consulta_prueba)
    resultados.append(respuesta)

# Calcular similitud sem√°ntica entre respuestas
consistencia_score = calcular_similitud_promedio(resultados)
```

#### 2. **Similitud Sem√°ntica**

Se utilizan embeddings para medir la similitud entre respuestas a consultas id√©nticas:

```python
from langchain.embeddings import OpenAIEmbeddings
from sklearn.metrics.pairwise import cosine_similarity

embeddings_model = OpenAIEmbeddings()

# Convertir respuestas a embeddings
embeddings = [embeddings_model.embed_query(r) for r in resultados]

# Calcular similitud coseno entre pares
similitudes = []
for i in range(len(embeddings)):
    for j in range(i+1, len(embeddings)):
        sim = cosine_similarity([embeddings[i]], [embeddings[j]])[0][0]
        similitudes.append(sim)

consistencia_score = np.mean(similitudes)
# Umbral aceptable: consistencia_score >= 0.85
```
